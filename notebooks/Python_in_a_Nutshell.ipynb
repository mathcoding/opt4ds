## 1.3 Flow Control
In the previous cells, we used only small blocks of code to demonstrate basic data structures. When writing a simple function or a larger program, it is necessary to write statements that control the flow of execution. Here we review the syntax for flow control statements available in Python.

### 1.3.1 Conditionals
The first flow control statement is the `if ... else` construct, which has the following syntax:

```
if <logical condition>:
  # Execute the first block of code, e.g.:
  print('True')
else:
  # Or execute the second block of code, e.g.:
  print('False')
```
The logical condition can be any logical operation (e.g., $a > 3$) or a function returning a logical value. The `else` branch is optional and can be omitted.

For instance, if we have to check whether a number is even, we can write:

````python
a = 14
if a % 2 == 0:
    print("Number a={} is EVEN".format(a))
else:
    print("Number a={} is ODD".format(a))
````

If we need to control several alternatives, we can use one (or more) `elif` conditions.

````python
if a > 0:
    print("positive")
elif a < 0:
    print("negative")
else:
    print("zero")
````

### 1.3.4 Exceptions
An [exception](https://docs.python.org/3/tutorial/errors.html) breaks the normal execution flow of a program in exceptional cases, such as when encountering unplanned errors.

By running this notebook on your own, you have likely already encountered exceptions as a way for the Python interpreter to report errors.

````python
p = (1,2)
p[0] = "a"
````

In this case, the exception was **raised** by an internal function of the Python interpreter. 

In order to control the execution flow of a program, the exceptions can be **caught** using the following syntax:

````python
try:
    # Block of code that might be raise an exception
    p = (1,2)
    p[0] = "a"
except Exception as e:
    # Block of code executed in case of an exception
    print(e)
    print(type(e))
````

### 1.4.1 Pure Functions
You can think of computer *pure* functions as if they were mathematical functions, where the output depends only on the input (i.e., there are no side effects).

Consider the problem of computing the hypotenuse of a triangle: $c^2 = a^2 + b^2$. We can write a small **function** that computes the square of the hypotenuse length $c$, given the two values of $a$ and $b$:

````python
def SquareHypotenuse(a, b):
    c = a**2 + b**2
    return c
````

In the previous example, we used the following syntax:

```
def <FunctionName>(<list of formal parameters, comma-separated>):
  <block of code>
```

There are a few points to note:

1. The keyword `def` is reserved and is used to define a function.
2. **FunctionName** is any name you want to use to define your function.
3. The parentheses and the colon are mandatory.
4. The **list of formal parameters** is up to you, and it can be of any (reasonable) size.
5. **MOST IMPORTANTLY:** the indentation after the `def` line is part of the **semantics** of the construct and denotes the scope of the function.
6. **Function DEFINITION** is different from **Function INVOCATION**. With the previous block, we defined the function `SquareHypotenuse`, but we have not yet passed any actual parameters to it, which would trigger the computation of the value `c`.

If we want to compute the square of the hypotenuse of a triangle with base equal to 4 and height equal to 3, we can write:

````python
b, h = 4, 3  # We are defining and unfolding the pair (4, 5) on the fly
hyp = SquareHypotenuse(b, h)
print(hyp)
````

Indeed, a large number of mathematical functions are already implemented in Python. For instance, all the basic mathematical functions are implemented in the [math](https://docs.python.org/3/library/math.html?highlight=math#module-math) library.

If you want to import a single function, for example the square root function [`sqrt()`](https://docs.python.org/3/library/math.html?highlight=math#math.sqrt), you can write:

````python
from math import sqrt
print(sqrt(2))
````

At this point, it should be easy to write a function that computes the actual value of the hypotenuse (and not its square) by using function composition.

````python
def Hypotenuse(a, b):
    return sqrt(SquareHypotenuse(a,b))
````

````python
print(Hypotenuse(4,3))
````

Indeed, functions can be chained several times.

````python
c = Hypotenuse(Hypotenuse(3,4), 5+2*SquareHypotenuse(2, sqrt(4)))
print(c)
````

### 1.4.2 Procedures
Procedures are like functions, but they do not explicitly return any value.

````python
def Print(a, b, c=77):
    print("a = {}, b = {}, c = {}".format(a, b, c))
````

````python
Print(2,3)
````

A typical error is to use a procedure as if it were a pure function. The syntax is correct, but it misuses the procedure.

````python
d = Print(a, b)
print(d)
````

Notice that there is no `return` keyword in the procedure `Print(a, b, c)`. However, in this case, it is as if there were a `return None` statement at the end of the procedure definition. For this reason, procedures are also called *void functions*.

````python
def Print(a, b, c=77):
    print("a = {}, b = {}, c = {}".format(a, b, c))
    return None
````

````python
print(Print(1,1,2))
````

````python
None
````

### 1.4.4 Map and Filter
Among the Python built-in functions, we like to stress the importance of [map](https://docs.python.org/3/library/functions.html#map) and [filter](https://docs.python.org/3/library/functions.html#filter).

They are higher-order functions; that is, functions that take other functions as input.

The `map` function takes as input a function *f* and any data type that is a sequence (an **iterable**, to be precise), and it applies the function *f* to every element of the sequence.

The `filter` function takes as input a predicate *p* and any data type that is a sequence, and it returns a new sequence containing only the elements for which the predicate *p* returns `True`.

For example:

````python
Ls = [1, 2, 3, 4, 5]   # String, list, tuple, dictionaries are iterable
for e in map(lambda x: x*x, Ls):
    print(e)
````

````python
Ls = [1, 2, 3, 4, 5]   # String, list, tuple, dictionaries are iterable
for e in filter(lambda x: x%2 == 0, Ls):
    print(e)
````

## 1.5 Classes and Objects
In Sections 1.1 and 1.2, we reviewed basic data types (numbers and strings) and structured data types (`tuple`, `list`, and dictionaries).

In this section, we show how we can define new data types, which can be composed of other data. New data types are defined using the syntax for defining `classes`. You can think of `class` as a synonym for `type`.

### 1.5.1 Syntax
The minimal syntax to define a new type of data is the following:

```
class <NameOfClass>(object):
  
  def __init__(self):
    self.<nameOfVariable> = <value>
    ...
    
  def __str__(self):
    s = <convert the data type to a representative string>
    return s
```

The term *class* is used here as a synonym for **type**, and it informs the interpreter that, in the following lines of code, we are declaring a new data type. The basic methods that every class should have are:

1. The constructor method `__init__(self)`, which is used to initialize the internal data of the class. The method can take one or more parameters as input, as we will see in the next example.

2. The string conversion method `__str__(self)`, which returns a string that gives a human-readable representation of the actual values of the class attributes.
